```python
#Tratar Valores Nulos, para numeros
for col in dados.select_dtypes(include='number'):
    dados[col].fillna(dados[col].mean(), inplace=True)
```

```python
dados
```

```python
#Para texto, substitua nulos pela moda
for col in dados.select_dtypes(include='object'):
    dados[col].fillna(dados[col].mode()[0], inplace=True)
```

```python
# Verificar valores nulos por coluna
print("Valores nulos por coluna:")
print(dados.isnull().sum())
```

```python
dados
```

```python
#Ajustar Colunas de Data
#dados['data_inversa'] = pd.to_datetime(dados['data_inversa'], format='%d/%m/%Y', errors='coerce')
dados['data_inversa'] = pd.to_datetime(dados['data_inversa'], dayfirst=True, errors='coerce')


```

```python
# Lista de colunas para ajustar
colunas_para_corrigir = ['km', 'latitude', 'longitude']  # Adicione as colunas problemáticas

# Substituir vírgulas por pontos e converter para float
for coluna in colunas_para_corrigir:
    dados[coluna] = dados[coluna].str.replace(',', '.', regex=True)  # Trocar vírgulas por pontos
    dados[coluna] = pd.to_numeric(dados[coluna], errors='coerce')  # Converter para float
```

```python
print(dados['data_inversa'].isnull().sum())
```

```python
# Identificar linhas com NaT
linhas_invalidas = dados[dados['data_inversa'].isnull()]

print("Linhas com datas inválidas:")
print(linhas_invalidas)


```

```python
# Remover linhas com NaT na coluna 'data_inversa'
dados.dropna(subset=['data_inversa'], inplace=True)

# Verificar o resultado
dados
```

#Engenharia de Atributos

```python
#Criar Informações Derivadas de Datas
#Adicione colunas como dia da semana, mês e ano:

dados['mes'] = dados['data_inversa'].dt.month
dados['ano'] = dados['data_inversa'].dt.year
```

```python
dados.head()
```

```python
#Categorizar Períodos do Dia, criar uma coluna para classificar os horários:

def periodo_do_dia(hora):
    if 0 <= hora < 6:
        return 'MADRUGADA'
    elif 6 <= hora < 12:
        return 'MANHÃ'
    elif 12 <= hora < 18:
        return 'TARDE'
    else:
        return 'NOITE'

dados['periodo_dia'] = dados['horario'].str[:2].astype(int).apply(periodo_do_dia)
```

```python
dados.head()
```

```python
#calcular total de feridos
dados['total_feridos'] = dados['feridos_leves'] + dados['feridos_graves']
```

#Analise Exploratoria

```python
import matplotlib.pyplot as plt

# Agrupar por tipo de acidente e somar os feridos
feridos_por_tipo = dados.groupby('tipo_acidente')['total_feridos'].sum().sort_values()

# Criar o histograma
plt.figure(figsize=(10, 6))
feridos_por_tipo.plot(kind='bar', color='skyblue', title='Total de Feridos por Tipo de Acidente')
plt.xlabel('Tipo de Acidente')
plt.ylabel('Total de Feridos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
