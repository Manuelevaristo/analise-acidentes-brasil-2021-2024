#importar bibliotecas

```python
import pandas as pd


```

#Tratar os arquivos

```python
# Abrir o arquivo e substituir delimitadores
with open("dados/2022_corrigido.csv", "r") as file:
    content = file.read().replace(",", ";")  # Substituir ',' por ';'

# Salvar o conteúdo padronizado
with open("dados/2022_padronizado.csv", "w") as file:
    file.write(content)
```

#carregar os dados e eliminar colunas desnecessárias

```python

# Carregar o CSV original
dados = pd.read_csv("dados/2022_padronizado.csv", delimiter=";", quotechar='"')

# Manter apenas as primeiras 30 colunas
dados = dados.iloc[:, :30]

# Salvar o arquivo atualizado
dados.to_csv("dados/2022_atualizados.csv", sep=";", index=False)

# Exibir informações para verificar
dados.head()
```

#carregando dados organizados

```python

dados_2021=pd.read_csv("dados/2021.csv", delimiter=';', encoding='latin1')
dados_2022=pd.read_csv("dados/2022_atualizados.csv", delimiter=';', encoding='latin1')
dados_2023=pd.read_csv("dados/2023.csv", delimiter=';', encoding='latin1')
dados_2024=pd.read_csv("dados/2024_atualizados.csv", delimiter=';', encoding='latin1')



```

#Consolidar e Salvar os dados

```python
# Concatene os DataFrames após tratamento
dados_consolidados = pd.concat([dados_2021, dados_2022, dados_2023, dados_2024], ignore_index=True)

# Salve o arquivo final
dados_consolidados.to_csv("dados/dados_consolidados.csv", index=False)
dados_consolidados
```

#Validação e Limpeza dos Dados

```python
dados
```

```python
dados.dtypes
```

```python
# Verificar valores nulos por coluna
print("Valores nulos por coluna:")
print(dados.isnull().sum())

    

```

```python
# Mostrar amostras de linhas com valores nulos
print("Linhas com valores nulos:")
print(dados[dados.isnull().any(axis=1)].head())
```

```python
# Contar duplicatas no DataFrame
num_duplicatas = dados.duplicated().sum()
print(f"Número de duplicatas: {num_duplicatas}")
```

```python
# Listar todas as colunas do DataFrame
print("Colunas no DataFrame:")
print(dados.columns)
```

```python
#Tratar Valores Nulos, para numeros
for col in dados.select_dtypes(include='number'):
    dados[col].fillna(dados[col].mean(), inplace=True)
```

```python
dados
```

```python
#Para texto, substitua nulos pela moda
for col in dados.select_dtypes(include='object'):
    dados[col].fillna(dados[col].mode()[0], inplace=True)
```

```python
# Verificar valores nulos por coluna
print("Valores nulos por coluna:")
print(dados.isnull().sum())
```

```python
dados
```

```python
#Ajustar Colunas de Data
#dados['data_inversa'] = pd.to_datetime(dados['data_inversa'], format='%d/%m/%Y', errors='coerce')
dados['data_inversa'] = pd.to_datetime(dados['data_inversa'], dayfirst=True, errors='coerce')


```

```python
dados
```

```python
print(dados['data_inversa'].isnull().sum())
```

```python
# Identificar linhas com NaT
linhas_invalidas = dados[dados['data_inversa'].isnull()]

print("Linhas com datas inválidas:")
print(linhas_invalidas)


```

```python
# Remover linhas com NaT na coluna 'data_inversa'
dados.dropna(subset=['data_inversa'], inplace=True)

# Verificar o resultado
dados
```
